# Data Description
The team at [FiveThirtyEight](http://fivethirtyeight.com/) became interested in answering some questions about *Star Wars* fans. In particular, they wondered: **does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?**

The team needed to collect data addressing this question. To do this, they surveyed *Star Wars* fans using the online tool SurveyMonkey. They received 835 total responses, which you download from their [GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

# Read in the data

```python
import pandas as pd

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
print(star_wars.head(10))
print(star_wars.columns)
```

```python
#print(star_wars["RespondentID"].value_counts)
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
# Remove any rows where RespondentID is NaN
print(star_wars["RespondentID"].value_counts)
```

# Cleaning and Mapping Yes/No Columns

```python
# print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]
#     .value_counts())
# print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]
#     .value_counts())

yes_no = {
    "Yes": True,
    "No": False
}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
star_wars.head()
```

# Cleaning and Mapping Checkbox Columns
There are several other columns containing answers to questions about the Star Wars movies. For some questions, the respondent had to check one or more boxes. This type of data is difficult to represent in columnar format.

```python
import numpy as np
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace":True,
    np.nan:False,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

cols = star_wars.columns[3:9]
for col in cols:
    star_wars[col] = star_wars[col].map(movie_mapping)

star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        }) #rename the columns

star_wars.head()
```

# Cleaning the Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

# Finding the Highest-Ranked Movie

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
star_wars.iloc[:,9:15].mean().plot.bar()
```

星战系列第5部似乎更受大家的认可，即《Star Wars: Episode V The Empire Strikes Back》。这与问题一致。

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

在所有接受采访的人中，星战系列第五部看过的人最多，大约有780人。第三部看过的人最少。

# Exploring the Data by Binary Segments
Now let's examine how certain segments of the survey population responded. There are several columns that segment our data into two groups.

```python
fans_y = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
fans_n = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] != True]
fans_y.head()
```

```python
fans_y.iloc[:,9:15].mean().plot.bar()
```

```python
fans_n.iloc[:,9:15].mean().plot.bar()
```

不管是不是星战系列的粉丝，均认为第五部是最好的。

```python
fans_y.iloc[:,3:9].sum().plot.bar()
```

```python
fans_n.iloc[:,9:15].mean().plot.bar()
```

粉丝中，第五部看的人数最多。而在不是粉丝的人群中，第三部观看的人数最多。
